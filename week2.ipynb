##Importing data and checking basic information

```python
import pandas as pd

# 데이터 파일 경로
cab_data_path = '/content/Cab_Data.csv'
city_data_path = '/content/City.csv'
customer_id_path = '/content/Customer_ID.csv'

# 데이터 불러오기
cab_data = pd.read_csv(cab_data_path)
city_data = pd.read_csv(city_data_path)
customer_id_data = pd.read_csv(customer_id_path)

# 기본 정보 확인
print(cab_data.info())
print(city_data.info())
print(customer_id_data.info())
```

##Data navigation and basic statistics

```python
# 기초 통계량 확인
print(cab_data.describe())
print(city_data.describe())
print(customer_id_data.describe())
```

##Visualize data


```python
import matplotlib.pyplot as plt
import seaborn as sns

company_avg_fare = cab_data.groupby('Company')['Price Charged'].mean().reset_index()
sns.barplot(x='Company', y='Price Charged', data=company_avg_fare)
plt.title('Average Fare by Taxi Company')
plt.show()
```

##Verification of the difference in taxi distance by taxi company

```python

company_avg_distance = cab_data.groupby('Company')['KM Travelled'].mean().reset_index()
sns.barplot(x='Company', y='KM Travelled', data=company_avg_distance)
plt.title('Average Travel Distance by Taxi Company')
plt.show()
```
